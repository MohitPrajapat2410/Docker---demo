```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Optimized").getOrCreate()
```

```python
data = [
    (1, "Alice", "Engineering", 75000, 25),
    (2, "Bob", "Marketing", 60000, 30),
    (3, "Charlie", "Engineering", 80000, 35),
    (4, "David", "Sales", 65000, 28),
    (5, "Eve", "Marketing", 70000, 32)
]
columns = ["Id", "Name", "Department", "Salary", "Age"]
df = spark.createDataFrame(data, columns)
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import StringType, IntegerType
import time
optimized_df = df.filter(col("Salary") > 500) \
                 .filter(col("Id") < 5000000) \
                 .select("Id", "Salary")

optimized_df.show()
```

```python
optimized_df.explain()
```

```python
start_time = time.time()
count_uncatched = optimized_df.count()
end_time = time.time()
print(f"1. Optimized Execution | count: {count_uncatched} | Time : {round(end_time - start_time,4)}")
```

```python
optimized_df.cache()
```

```python
start_time = time.time()
count_catched = optimized_df.count()
end_time = time.time()
print(f"1. Optimized Execution | count: {count_catched} | Time : {round(end_time - start_time, 4)} seconds")
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,udf
from pyspark.sql.types import StringType, IntegerType
import time

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("UDFDemo") \
    .getOrCreate()

print(type(spark))
```

```python
data = [("alice",25),("Mohit",19),("Nikhil",22)]

columns = ["name","age"]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
def categorize_age(age):
    if age >= 18:
        return "Adult"
    return "Minor"
```

```python
# Create UDF
age_category_udf = udf(categorize_age, StringType())

# Apply UDF to DataFrame
df_method1 = df.withColumn(
    "Category",
    age_category_udf(col("age"))
)
print("Method 1: Standard UDF via DataFrame API")
df_method1.show()
```

```python
# For SQL
spark.udf.register(
    "sql_categorize_age",
    categorize_age,
    StringType()
)

df.createOrReplaceTempView("people")
```

```python
sql_df = spark.sql("""
    SELECT
        Name,
        Age,
        sql_categorize_age(Age) AS Category
    FROM people
""")

sql_df.show()
```

```python
def can_drive(age):
    if age >=18:
        return "can drive"
    return "can not drive"
    
```

```python
can_drive_udf = udf(can_drive, StringType())

# Apply UDF to DataFrame
df_method2 = df.withColumn(
    "drive",
    can_drive_udf(col("age"))
)
print("Method 2: Standard UDF via DataFrame API")
df_method2.show()
```
