```python
from pyspark.sql import SparkSession
```

```python
# 1. Initialize SparkSession
spark = SparkSession.builder \
 .appName("PartitioningDemo") \
 .getOrCreate()
```

```python
# 2. Create a sample DataFrame
# (Using range creates a quick dataset of 1 million rows)
df = spark.range(0, 1000000)
```

```python
# 3. Print BEFORE partitioning
initial_partitions = df.rdd.getNumPartitions()
print("=========================================")
print(f"BEFORE: Initial number of partitions: {initial_partitions}")
print("=========================================")
```

```python
# 4. Apply REPARTITION (Used to increase or evenly distribute partitions -causes a full shuffle)
df_repartitioned = df.repartition(10)

# Print AFTER repartition
repartition_count = df_repartitioned.rdd.getNumPartitions()
print(f"AFTER REPARTITION: Partitions changed to: {repartition_count}")
print("=========================================")
```

```python
# 5. Apply COALESCE (Used to decrease partitions efficiently - minimizes data shuffling)
df_coalesced = df_repartitioned.coalesce(2)
# Print AFTER coalesce
coalesce_count = df_coalesced.rdd.getNumPartitions()
print(f"AFTER COALESCE: Partitions reduced to: {coalesce_count}")
print("=========================================")
```

```python
# Optional: Stop the Spark session when done
spark.stop()
```
